```python
import numpy as np
import matplotlib.pyplot as plt
import librosa as lb
import librosa.display 
from matplotlib.pyplot import specgram
import os
import IPython.display as ipd
```

```python
#REVDASS , emotion:angry , gender:female
revdass="E:\\Kaggle\\Audio Emo Classifier\\RAVDESS\\Actor_08\\03-01-05-02-01-01-08.wav"
data,s_rate=lb.load(revdass,sr=44100,offset=0.5,duration=2.5,res_type='kaiser_fast')
mfcc_f=lb.feature.mfcc(data,sr=s_rate,n_mfcc=13)

plt.figure(figsize=(10,10))
plt.subplot(3,1,1)
librosa.display.specshow(mfcc_f,x_axis='time')
plt.ylabel('MFCC')
plt.colorbar()

ipd.Audio(revdass)
```

```python
#Compare the Angry female and Angry male for the same sentence uttered.
revdass_female="E:\\Kaggle\\Audio Emo Classifier\\RAVDESS\\Actor_08\\03-01-05-02-01-01-08.wav"
revdass_male="E:\\Kaggle\\Audio Emo Classifier\\RAVDESS\\Actor_09/03-01-05-01-01-01-09.wav"

X, sample_rate = librosa.load(revdass_female, res_type='kaiser_fast',duration=2.5,sr=44100,offset=0.5)  
#female = librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=13)
female = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=13), axis=0)
print(len(female))

X, sample_rate = librosa.load(revdass_male, res_type='kaiser_fast',duration=2.5,sr=44100,offset=0.5)  
#male = librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=13)
male = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=13), axis=0)
print(len(male))

plt.figure(figsize=(20,20))
plt.subplot(2,1,1)
plt.plot(female,label='Female')
plt.plot(male,label='Male')
plt.title('The females have a higher pitch. Period.')
plt.legend()
```
